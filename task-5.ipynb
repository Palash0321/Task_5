#### Converting 'Order Date' to datetime objects

```python
orders_df['Order Date'] = pd.to_datetime(orders_df['Order Date'], format='%d-%m-%Y')
```

```python
orders_df
```

#### Extracting temporal components

```python
orders_df['Order Year'] = orders_df['Order Date'].dt.year
```

```python
orders_df
```

```python
orders_df['Order Month'] = orders_df['Order Date'].dt.month
```

```python
orders_df
```

```python
orders_df['Order Month Name'] = orders_df['Order Date'].dt.month_name()
orders_df
```

```python
orders_df['Order Day'] = orders_df['Order Date'].dt.day
orders_df
```

```python
orders_df['Order Day of Week'] = orders_df['Order Date'].dt.day_name()
orders_df
```

#### Orders Over Time (Yearly Trend)

```python
print("\n--- Visualizing Yearly Order Trend ---")
yearly_orders = orders_df.groupby('Order Year').size().reset_index(name='Order Count')
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_orders, x='Order Year', y='Order Count', marker='o')
plt.title('Total Orders by Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.xticks(yearly_orders['Order Year'].unique()) # Ensure all years are shown as ticks
plt.grid(True)
plt.tight_layout()
plt.show()
```

Observation: The number of orders seems to be high in 2018.


#### Orders Over Time (Monthly Trend - overall)

```python
print("\n--- Visualizing Monthly Order Trend (Overall) ---")
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_orders = orders_df.groupby('Order Month Name').size().reindex(month_order).reset_index(name='Order Count')
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_orders, x='Order Month Name', y='Order Count', hue='Order Month Name', palette='viridis')
plt.title('Total Orders by Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Observation: This will show which months are most active.


#### Top 10 States by Order Count



```python
print("\n--- Visualizing Top 10 States by Order Count ---")
top_states = orders_df['State'].value_counts().nlargest(10).reset_index(name='Order Count')
top_states.columns = ['State', 'Order Count'] # Rename columns for clarity
plt.figure(figsize=(12, 7))
sns.barplot(data=top_states, x='Order Count', y='State', hue='State',legend=False, palette='mako')
plt.title('Top 10 States by Order Count')
plt.xlabel('Number of Orders')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

Observation: This will highlight the geographical distribution of orders by state.


#### Top 10 Cities by Order Count

```python
print("\n--- Visualizing Top 10 Cities by Order Count ---")
top_cities = orders_df['City'].value_counts().nlargest(10).reset_index(name='Order Count')
top_cities.columns = ['City', 'Order Count'] # Rename columns for clarity
plt.figure(figsize=(12, 7))
sns.barplot(data=top_cities, x='Order Count', y='City',hue='City',legend=False, palette='rocket')
plt.title('Top 10 Cities by Order Count')
plt.xlabel('Number of Orders')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

Observation: This will show the top-performing cities.


#### Top 10 Customers by Order Count

```python
print("\n--- Visualizing Top 10 Customers by Order Count ---")
top_customers = orders_df['CustomerName'].value_counts().nlargest(10).reset_index(name='Order Count')
top_customers.columns = ['CustomerName', 'Order Count'] # Rename columns for clarity
plt.figure(figsize=(12, 7))
sns.barplot(data=top_customers, x='Order Count', y='CustomerName', hue='CustomerName', legend=False, palette='crest')
plt.title('Top 10 Customers by Order Count')
plt.xlabel('Number of Orders')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```

Observation: This will identify the most loyal/frequent customers.


#### Orders by Day of Week


```python
print("\n--- Visualizing Orders by Day of Week ---")
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_orders = orders_df.groupby('Order Day of Week').size().reindex(day_order).reset_index(name='Order Count')
plt.figure(figsize=(10, 5))
sns.barplot(data=daily_orders, x='Order Day of Week', y='Order Count',hue='Order Count', legend=False, palette='cubehelix')
plt.title('Total Orders by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

Observation: This will show daily patterns in order placement.
